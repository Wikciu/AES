Sprawdzam co fajnego jest w pliku

```python
titanic_df = pd.read_csv(base_path / 'titanic.csv', index_col='PassengerId')
titanic_df.info()
titanic_df.values[1]
```

Niektóre dane są puste i nie mają wpływu na szukany wynik

```python

titanic_df["Age"].fillna(titanic_df["Age"].median(), inplace=True)

titanic_df["Embarked"].fillna(titanic_df["Embarked"].mode()[0], inplace=True)

titanic_df["Cabin"].fillna("Unknown", inplace=True)
```

```python
# Dodaje ceche klasy
titanic_df["Klasa"] = titanic_df["Pclass"].astype(int)
# Dodaje ceche płci
titanic_df["Plec"] = titanic_df["Sex"].map({"male": 0, "female": 1})
```

Usuwam imię, numer biletu, kabine, małżeństwa, dzieci

```python
Usundrop = ["Name", "Ticket", "Cabin", "SibSp", "Parch"]
titanic_df.drop(columns=Usundrop, inplace=True)
```

```python
titanic_df.info()
# Po zmianach
```

```python
#Wykres zależny od płci
plt.figure(); titanic_df.groupby("Plec")["Survived"].mean().sort_index().plot(kind="bar")
plt.title("Zależność płci (0=chłop, 1=baba)"); plt.xlabel("Plec"); plt.ylabel(""); plt.show()
```

```python
plt.figure(); titanic_df.groupby("Klasa")["Survived"].mean().sort_index().plot(kind="bar")
plt.title("Przeżycie według Klasy"); plt.xlabel("Klasa"); plt.ylabel("Średnia Ocalałych"); plt.show()
```

```python
#Porównanie wybranych cech
pivot = titanic_df.pivot_table(index="Klasa", columns="Plec", values="Survived", aggfunc="mean")
print(pivot)
```
